7. Рассмотрим случайную матрицу  $15 × 2 $, представляющую декартовы координаты, преобразуйте их в полярные координаты.

```python
coords = np.random.rand(15, 2) * 10

x = coords[:, 0]
y = coords[:, 1]

r = np.sqrt(x**2 + y**2)

theta = np.arctan2(y, x)

polar_coords = np.column_stack((r, theta))

print(coords)
print(polar_coords)
```

1. Сколько спортсменов участвовали в Летних Олимпийских играх за всё время?

```python


olympics_data = pd.read_csv('olympics_dataset.csv')

summer_olympics = olympics_data[olympics_data['Season'] == 'Summer']

unique_athletes_count = summer_olympics['Name'].nunique()

unique_athletes_count
```

2. Найдите участника с наибольшим количеством медалей

```python
medalists = olympics_data[olympics_data['Medal'] != 'No medal']

medal_counts = medalists.groupby('Name').size()

top_medalist = medal_counts.idxmax()
top_medal_count = medal_counts.max()

str(top_medalist) + " " + str(top_medal_count)
```

3. Какой спорт является самым популярным, то есть в нём участвовало больше всего спортсменов за всё время?

```python
sport_popularity = olympics_data.groupby('Sport')['Name'].nunique()

most_popular_sport = sport_popularity.idxmax()
most_popular_count = sport_popularity.max()

str(most_popular_sport) + " " +  str(most_popular_count)
```

4. Кто был победителем медального зачёта в 2004 году?

```python
medals_2004 = olympics_data[(olympics_data['Year'] == 2004) & (olympics_data['Medal'] != 'No medal')]

country_medal_count = medals_2004.groupby('Team').size()

top_country = country_medal_count.idxmax()
top_medal_count = country_medal_count.max()

str(top_country) + " " +  str(top_medal_count)
```

5. В каком виде спорта США завоевало наибольшее число медалей?

```python
us_medals = olympics_data[(olympics_data['Team'] == 'United States') & (olympics_data['Medal'] != 'No medal')]
us_sport_medals = us_medals.groupby('Sport').size()
top_us_sport = us_sport_medals.idxmax()
top_us_medal_count = us_sport_medals.max()

str(top_us_sport) + " " +  str(top_us_medal_count)
```

# С помощью графика покажите:

6. Сколько женщин и мужчин участвовало в Летних Олимпийских играх каждый год?

```python
import matplotlib.pyplot as plt

summer_data = olympics_data[olympics_data['Season'] == 'Summer']

gender_participation = summer_data.groupby(['Year', 'Sex']).size().unstack()

gender_participation.plot(kind='bar', figsize=(15, 7), width=0.8)

plt.title('Количество женщин и мужчин, участвовавших в Летних Олимпийских играх (по годам)')
plt.xlabel('Год')
plt.ylabel('Количество участников')
plt.legend(title='Пол', labels=['Мужчины', 'Женщины'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

7. Какие 10 стран на текущий момент имеют наибольшее количество золотых медалей?


```python
gold_medals = olympics_data[olympics_data['Medal'] == 'Gold']

gold_medal_counts = gold_medals.groupby('Team').size().sort_values(ascending=False)

top_10_countries = gold_medal_counts.head(10)

top_10_countries.plot(kind='bar', figsize=(12, 6), color='gold')

plt.title('Топ-10 стран с наибольшим количеством золотых медалей')
plt.xlabel('Страна')
plt.ylabel('Количество золотых медалей')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

8. Сколько золотых, серебряных и бронзовых медалей зарабатывал СССР отдельно по каждым играм?

```python
ussr_medals = olympics_data[(olympics_data['Team'] == 'Russia') & (olympics_data['Medal'] != 'No medal')]

ussr_medals_count = ussr_medals.groupby(['Year', 'Medal']).size().unstack(fill_value=0)

ussr_medals_count.plot(kind='bar', figsize=(15, 7), width=0.8)

plt.title('Количество золотых, серебряных и бронзовых медалей СССР по годам')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.legend(title='Тип медали', labels=['Золото', 'Серебро', 'Бронза'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

9. Сколько медалей завоёвывали спортсмены и спортсменки (отдельно) команды Китая на каждой олимпиаде?


```python
china_medals = olympics_data[(olympics_data['Team'] == 'China') & (olympics_data['Medal'] != 'No medal')]

china_gender_medals = china_medals.groupby(['Year', 'Sex']).size().unstack(fill_value=0)

china_gender_medals.plot(kind='bar', figsize=(15, 7), width=0.8)

plt.title('Количество медалей спортсменов и спортсменок Китая по олимпиадам')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.legend(title='Пол', labels=['Мужчины', 'Женщины'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
